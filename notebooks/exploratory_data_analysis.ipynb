# EXPLORATORY DATA ANALYSIS 

## Introduction 
This notebook performs Exploratory Data Analysis (EDA) on the cleaned Chennai housing dataset.
The objective is to understand the distribution of house prices, identify key features such as area, location, resale status, and amenities, and make informed decisions for feature engineering and modeling in the REIS project

### 1. Import Libraries & Load Dataset

```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_chennai_dataset.csv")

df.head()
```

### 2. Data Overview

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include="all")
```

### 3. Target Variable Analysis: Price
#### a) Distribution of Price

```python
plt.hist(df["price"], bins=50)
plt.xlabel("price")
plt.show()
```

- Price is highly right-skewed
- Most properties are in the lower price range, with a few very expensive houses

#### b) Skewness & Central Tendency

```python
df["price"].mean(), df["price"].median()  
```

```python
df["price"].skew(),df["price"].kurtosis()
```

- Most house prices are on the lower side, with a few very expensive properties.
- Because of these expensive houses, the average (mean) price is much higher than the typical price
- The price distribution has extreme values (outliers) on the higher end
- The median price represents the market better than the average price

### 4. Target Variable Adjustment
#### a) Log(price) Distribution

```python
plt.hist(np.log(df["price"]), bins=50)
plt.xlabel("log(price)")
plt.show()
```

- After log transformation, the price distribution becomes much more balanced
- Extreme prices have less influence
- Most values now lie in a compact, bell-shaped range

#### b) Create log_price

```python
df["log_price"] = np.log(df["price"])
```

### 5. Numerical Feature vs Target Analysis
#### a) Area vs log(price)

```python
plt.figure(figsize=(15,5))
plt.scatter(x=df["area_sqft"],y=df["log_price"])
plt.xlabel("Area (sqft)")
plt.ylabel("log (price)")
plt.title("Scatter plot of AREA vs LOG PRICE")
plt.show()
```

- Price generally increases as area increases
- The relationship is positive but not perfectly linear
- Smaller houses show wide price variation
- Very large houses are expensive, but not proportionally 

#### b) Bedrooms vs log(price)

```python
sns.boxplot(x="no_of_bedrooms",y="log_price",data=df)
plt.xlabel("No of Bedrooms")
plt.ylabel("log (price)")
plt.title("Box plot of No. of bedrooms vs log (price)")
plt.show()
```

- Price increases as number of bedrooms increases
- Price jump from 1  2  3 bedrooms is clear
- For 45 bedrooms, price variation is high, indicating influence of location and size

### 6. Categorical Feature vs Target Analysis
#### a) Resale vs log(price)

```python
sns.boxplot(x="resale", y="log_price", data=df)
plt.xlabel("Resale (0 = New, 1 = Resale)")
plt.ylabel("log(price)")
plt.title("Resale vs log(price)")
plt.show()
```

- New properties often show slightly higher median prices
- There is a large overlap between new and resale prices
- Resale status has some impact, but it is not a strong driver of price

#### b) Location vs Median Price

```python
location_stats = (df.groupby("location")["log_price"].agg(
                  count="count",median_log_price="median").sort_values("count", ascending=False))

location_stats.head()
```

```python
rare_locations = (location_stats[location_stats["count"] < 20].index)

df["location_grouped"] = df["location"].apply(lambda x: "other" if x in rare_locations else x)

df["location_grouped"].value_counts().head()
```

```python
loc_grouped_stats = (df.groupby("location_grouped")["log_price"].agg(
                     count="count",median_log_price="median").sort_values("median_log_price", ascending=False))

loc_grouped_stats.head(10)
```

```python
loc_grouped_stats.loc["other"]
```

```python
plt.figure(figsize=(15,5))
plt.bar(loc_grouped_stats.index, loc_grouped_stats["median_log_price"])
plt.xlabel("Location (grouped)")
plt.ylabel("Median log(price)")
plt.title("Median log(price) by location_grouped")
plt.xticks(rotation=90)
plt.show()
```

- Location has a strong influence on house prices
- Premium areas (e.g., Raja Annamalai Puram, Egmore) have clearly higher median prices
- Grouping rare locations into Other avoids noisy and unreliable comparisons
- This justifies using location_grouped for modeling

#### c) Amenity vs log (price)

```python
amenity_cols = ["nearby_school_1km","nearby_hospital_1km","nearby_mall_1km","nearby_gym_1km",
                "swimmingpool","rainwaterharvesting","sportsfacility","powerbackup","carparking",
                "childrens_playarea","liftavailable"]

df["amenity_score"] = df[amenity_cols].sum(axis=1)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["amenity_score"], df["log_price"], alpha=0.3)
plt.xlabel("Amenity Score")
plt.ylabel("log(price)")
plt.title("Amenity Score vs log(price)")
plt.show()
```

- The relationship is weak but positive
- Houses with more amenities show less price uncertainty
- Amenities improve perceived quality but do not fully determine price

### 7. Correlation Analysis

```python
corr = df[["log_price","area_sqft","no_of_bedrooms","amenity_score","resale"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation heatmap (selected features)")
plt.show()
```

- Property size features show the strongest association with price.
- Area and number of bedrooms are highly correlated, reflecting structural dependence.
- Amenity score contributes weak but independent signal.
- Resale status shows minimal linear relationship with price.
- No severe multicollinearity is observed among the selected features.

## Summary
- House prices are highly right-skewed; median is more reliable than mean
- Log transformation stabilizes price distribution and is used for modeling
- Location strongly influences price; rare locations were grouped as Other"
- Larger area and more bedrooms generally increase price
- Amenities show a positive but moderate impact when aggregated
